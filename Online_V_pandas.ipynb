# Function Application and Mapping

```python
import numpy as np
table=pd.DataFrame(np.random.randn(4,3), columns=list('tre'),index=['1st','2nd','3rd','4th',])
print(table['r'].min())
print(table['r'].min()-table['t'].max())

f= lambda x: x.max()-x.min()
n=table.apply(f)
print(n,type(n))
display('---')
n=table.apply(f,axis=1)
print(n)
```

```python
def min_max(x):
    
    return pd.Series( [x.min(),x.max()], index=['min','max'] )
n=table.apply(min_max)
print(n,type(n))
```

# Working with Databases

```python
import sqlite3
#creating table
query = """CREATE TABLE test(a VARCHAR(20),b VARCHAR(20),c REAL, d INTEGER);"""
con = sqlite3.connect('mydata.sqlite')
con.execute(query)
```

```python
#inserting data
data = [('Atlanta', 'Georgia',1.25,6),
        ('Tallahassee','Florida', 2.6,3),
        ('Sacramento','California',1.7,5)
       ]
stmt = "INSERT INTO test VALUES( ?, ?, ?, ?)"
con.executemany(stmt, data)
con.commit()
```

```python
#selecting extracting data
cursor = con.execute('select * from test')
rows = cursor.fetchall()
print(type(rows), rows)
```

```python
#print(cursor.description)
data_df = pd.DataFrame(rows,columns=[x[0] for x in cursor.description])
print(data_df,type(data_df))
```

```python
#conda install -c anaconda sqlalchemy
#using sqlalchemy
import sqlalchemy as sqla
db = sqla.create_engine('sqlite:///mydata.sqlite')
data_df = pd.read_sql('select * from test', db)
display(data_df)
```

# Handling Missing Data

```python
import pandas as pd
from numpy import nan as NA
data ={'a':[10,NA,4,50,80],
       'b':[5,NA,12.5,NA,11],
       'c':[NA,NA,14.12,4,20],
       'd':[4,NA,NA,NA,5]}
set1 = pd.DataFrame(data)
display(set1)
drop = set1.dropna(axis=0)
display(drop)
drop = set1.dropna(how='all')
display(drop)
```

# Removing Duplicates

```python
import pandas as pd
data = pd.DataFrame({'k1':['one','two']*3+['two'],
                     'k2':[1,1,2,3,3,4,4]} )
display(data)
#data.duplicated()
data.drop_duplicates()

data['v1'] = ['one','two','one','four','one','six','two']
print(data)
data = data.drop_duplicates(['k1','v1'])
display(data)
```

# Replacing Values

```python
display(data)
data.iloc[:2,0]=np.nan
data.iloc[1:3,2]=np.nan
df = data.replace(np.nan,100)
print(df)
df = df.replace([100,'two'],[1.11,2])
display(df)
```
