# How do Filter rows of pandas DataFrame by column value

```python
booleans=[]
for duration in movies.duration:
    if duration >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
movies[booleans]
```

```python
movies[movies.duration >= 200]
```

```python
movies[movies.duration >= 200].title
```

```python
movies.loc[movies.duration >= 200,'genre']
```

# multiple filter criteria to pandas DataFrame

```python
movies[(movies.duration >=200) & (movies.genre =='Action')]
```

```python
movies[(movies.genre =='Drama') | (movies.genre =='Action')]
```

```python
df=pd.DataFrame({'ID':[100,200,150,50],'quality':['nice','good','very good','excellent']})
df
```
